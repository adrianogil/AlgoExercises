# Studying factors of N

```python
max_number = 10000

# Initialize a variable to store the sum of percentages
percentage_sum = 0

x = []
y = []
yp = []

for i in range(4, max_number):
    # print(i)
    x.append(i)
    total_factors = 0
    factors = []

    for j in range(1, i+1):
        if (i % j) == 0:
            total_factors += 1
            factors.append(j)
    percentage_sum += (total_factors / i)
    current_percentage_str = "%.3f" % ((100.0*percentage_sum)/(i-3),)
    y.append(total_factors)
    yp.append(percentage_sum/(i-3))
    print(f"N: {i} - Number of factors: {total_factors} ({current_percentage_str} %): {factors}")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def plot_data(x, y, title):
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.grid(True)
    plt.show()

# # Example usage:
# x = [1, 2, 3, 4, 5]
# y = [2, 3, 5, 7, 11]

plot_data(x, y, title="Number of factors of N")
```

```python
plot_data(x, yp, title="% of factors of N")
```
